```python
from google.colab import drive
drive.mount('/content/gdrive')
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('/content/gdrive/MyDrive/ComputerVision/twins.jpg',cv2.IMREAD_GRAYSCALE)
# plt.imshow(img, cmap='gray')
# plt.show()

x=img.shape[0]
y=img.shape[1]
data=np.zeros(256)
for i in range(int(x/2),x):
  for j in range(int(y/2),y):
    z=img[i][j]
    data[z]+=1

bar_width=0.5
xlocations = np.array(range(len(data))) + bar_width
plt.bar(xlocations,data)
plt.title('Image Analysis')
plt.ylabel('Times')
```
